```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel("Structured_Data.xlsx")
```

```python
● Get all Pokemons whose spawn rate is less than 5%
● Get all Pokemons that have less than 4 weaknesses
● Get all Pokemons that have no multipliers at all
● Get all Pokemons that do not have more than 2 evolutions
● Get all Pokemons whose spawn time is less than 300 seconds.
Note - spawn time format is "05:32”, so assume “minute: second” format and perform the analysis.
● Get all Pokemon who have more than two types of capabilities
```

```python
list(df[df.spawn_chance*100< 5].name) # all Pokemons whose spawn rate is less than 5%
```

```python
import ast
list(df[df.weaknesses.apply(lambda X: len(ast.literal_eval(X)) < 4 )].name) # all Pokemons that have less than 4 weaknesses
```

```python
#all Pokemons that have no multipliers at all
list(df[df.multipliers.apply(lambda X: X == '[0]')].name)
```

```python
#all Pokemons that do not have more than 2 evolutions
df_new=df.copy()
df_new.dropna(inplace=True)
df_new[df_new.next_evolution.apply(lambda X: len(ast.literal_eval(X)) <= 2)].name
```

```python
#all Pokemons whose spawn time is less than 300 seconds 
#00:00 also included in list assuming that they are not teaking any time for spawn
df[df.spawn_time.apply(lambda X: int(X.split(':')[0])*60+int(X.split(':')[1]) < 300)].name
```

```python
# all Pokemon who have more than two types of capabilities
df[df.type.apply(lambda X: len(ast.literal_eval(X)) > 2 )] # there is no one with more that two capabilities
```
